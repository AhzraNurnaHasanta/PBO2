```python
class buku:
  def __init__(self, judul, pengarang, tahun_terbit):
    self.judul = judul
    self.pengarang = pengarang
    self.tahun_terbit = tahun_terbit

def main():
  buku1 = buku("Laut Tengah", "Berliana Timberly", 2023)
  buku2 = buku("Laskar Pelangi", "Andrea Hirata", 2005)

  print("Buku 1 : ",buku1.judul, buku1.pengarang, buku1.tahun_terbit)
  print("Buku 2 : ",buku2.judul, buku2.pengarang, buku2.tahun_terbit)

if __name__ == "__main__":
  main()
```
