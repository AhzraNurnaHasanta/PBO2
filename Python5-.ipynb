```python
class handphone:
  def __init__(self, merek, penyimpanan, harga):
    self.merek = merek
    self.penyimpanan = penyimpanan
    self.harga = harga

def main():
  hp1 = handphone("Samsung", "128 GB", 10000000)
  hp2 = handphone("Xiaomi", "64 GB", 5000000)

  print("HandPhone 1")
  print(hp1.merek)
  print(hp1.penyimpanan)
  print(hp1.harga)
  print("")
  print("HandPhone 2")
  print(hp2.merek)
  print(hp2.penyimpanan)
  print(hp2.harga)

if __name__ == "__main__":
  main()
```
