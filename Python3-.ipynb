```python
class mobil:
  def __init__(self, merk, warna):
    self.merk = merk
    self.warna = warna

def main():
  mobil1 = mobil("Avanza", "Hitam")
  mobil2 = mobil("Inova", "Putih")

  print("Merek Mobil 1 : ",mobil1.merk)
  print("Warna Mobil 1 : ",mobil1.warna)
  print("Merek Mobil 2 : ",mobil2.merk)
  print("Warna Mobil 2 : ",mobil2.warna)

if __name__ == "__main__":
  main()
```
