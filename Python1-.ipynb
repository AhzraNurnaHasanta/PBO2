```python
class mahasiswa:
  def __init__(self, nama, nim, krs):
    self.nama = nama
    self.nim = nim
    self.krs = krs

def main():
  mhs = mahasiswa("Ahzra Nurna Hasanta", 230705054, "PBO, Kecerdasan Bisnis, Jaringan Komputer")

  print("Nama : ", mhs.nama)
  print("NIM : ", mhs.nim)
  print("KRS : ", mhs.krs)

if __name__ == "__main__":
  main()
```
